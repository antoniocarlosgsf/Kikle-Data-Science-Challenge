## 📊 Principais Drivers de Performance dos Criativos

Para entender quais atributos estão mais associados a campanhas de alto desempenho, foi criada uma variável binária (`high_performance`), definida a partir da mediana do Klike Score.

---

### 🔥 Atributos Criativos

#### 🎬 Hook (primeiros 3 segundos)

- Campanhas com hook apresentam **73% de probabilidade de alta performance**, contra apenas **25% sem hook**.
- Entre as campanhas de melhor desempenho:
  - **88% possuem hook**, enquanto apenas **18% das de pior desempenho** utilizam esse recurso.

👉 O uso de hook é o fator mais relevante associado ao sucesso do criativo.

---

#### 👤 Presença de rosto humano

- Campanhas com rosto apresentam **63% de probabilidade de alta performance**, contra **34% sem rosto**.

👉 A humanização do conteúdo tem impacto significativo no desempenho.

---

#### 🎯 Call-to-action (CTA)

- Campanhas com CTA apresentam **56% de probabilidade de alta performance**, contra **35% sem CTA**.

👉 O CTA contribui positivamente, mas com impacto menor que o hook.

---

#### 📝 Legendas

- Campanhas com legenda apresentam melhor desempenho em média.
- No entanto, o impacto é mais moderado em comparação com outros atributos.

---

### 📱 Formato e Estrutura do Conteúdo

#### 📐 Formato do vídeo

- Vídeos verticais apresentam o melhor desempenho médio.
- Formatos quadrados são intermediários.
- Formatos horizontais tendem a performar pior.

---

#### 🧾 Densidade de texto

- **Medium** → melhor performance (~65)
- **Low** → intermediário (~58)
- **High** → pior performance (~50)

👉 Alta densidade de texto prejudica significativamente o desempenho do criativo.

---

### 📈 Métricas de Performance

- **CTR (~0.50)** e **engagement rate (~0.22)** apresentam correlação positiva com o Klike Score.
- No entanto, **atributos criativos (hook, rosto, CTA)** apresentam maior impacto do que métricas de mídia.

👉 Isso indica que a qualidade criativa é um fator mais determinante do que apenas o desempenho de mídia.

---

### 🏆 Comparação: Top vs Bottom Performers

Comparando o quartil superior (top 25%) com o inferior (bottom 25%):

- Hook:
  - 88% (top) vs 18% (bottom)

- Presença de rosto:
  - 77% vs 31%

- CTR:
  - 0.10 vs 0.04

- ROAS:
  - 3.57 vs 1.60

- Engagement rate:
  - 0.033 vs 0.020

👉 Os criativos de melhor desempenho combinam elementos estruturais fortes (hook, rosto) com melhores métricas de engajamento.

---

### 🎯 Conclusão

Os resultados indicam que a estrutura do criativo — especialmente a presença de hook e elementos humanos — tem impacto direto e significativo no sucesso da campanha, frequentemente superando fatores relacionados ao investimento ou escala de mídia.

## 📊 Análise por Plataforma

Para aprofundar a análise exploratória, foram avaliadas as distribuições das principais métricas de performance por plataforma, bem como a interação entre atributos criativos e essas métricas.

---

### 📈 Diferenças gerais entre plataformas

#### 💰 ROAS
- **TikTok** apresenta a maior mediana e maior dispersão de ROAS.
- **Meta** possui performance intermediária.
- **LinkedIn** apresenta menor ROAS e menor variabilidade.

👉 O TikTok oferece maior potencial de retorno, porém com maior volatilidade, enquanto o LinkedIn tende a ser mais estável, mas com menor upside.

---

#### 💬 Engagement Rate
- **TikTok** possui níveis significativamente mais altos de engajamento.
- **Meta** e **LinkedIn** apresentam valores mais baixos e semelhantes.

👉 Isso indica que o TikTok é mais sensível à qualidade do criativo e favorece conteúdos que capturam atenção rapidamente.

---

### 🎬 Impacto de atributos criativos por plataforma

#### 🔥 Hook

- O uso de hook aumenta o **CTR em todas as plataformas**.
- O impacto é mais forte no **TikTok e LinkedIn**.

👉 O hook eleva o nível base de performance, sendo um dos principais fatores de captura de atenção.

---

- Em relação ao **engagement rate**:
  - O impacto é **muito forte no TikTok**.
  - Moderado no Meta.
  - Menor no LinkedIn.

👉 O efeito do hook é amplificado em plataformas de consumo rápido, como o TikTok.

---

- Em relação ao **watch time**:
  - O impacto é maior em **Meta e LinkedIn**.
  - Menor no TikTok.

👉 Em plataformas de consumo mais lento, o hook contribui mais para retenção.

---

#### 👤 Presença de rosto

- A presença de rosto aumenta o **CTR em todas as plataformas**.
- Também aumenta o **engagement rate**, especialmente no TikTok.

👉 Elementos humanos tornam o conteúdo mais atrativo e aumentam tanto a atenção quanto o engajamento.

---

#### 🎯 Call-to-Action (CTA)

- O CTA melhora o **CTR em todas as plataformas**.
- O efeito é mais evidente no TikTok.

👉 O CTA atua como um mecanismo de conversão de interesse em ação (clique).

---

#### 🔄 Conversões

- O hook aumenta conversões em média, porém com alta variabilidade.
- A dispersão indica que outros fatores (produto, oferta, público) também são determinantes.

👉 O impacto do criativo nas conversões é mais indireto, mediado por métricas como CTR e engagement.

---

### 🧾 Densidade de texto

- **Text density alta** apresenta pior desempenho em CTR e engagement.
- **Medium** apresenta melhor performance.
- O padrão é consistente em todas as plataformas.

👉 Excesso de texto prejudica a performance do criativo, possivelmente por aumentar a carga cognitiva do usuário.

---

### 🎯 Conclusão

A análise mostra que:

- Existem **padrões consistentes entre plataformas**, como:
  - impacto positivo de hook
  - impacto positivo de elementos humanos
  - impacto negativo de excesso de texto

- No entanto, a **intensidade desses efeitos varia significativamente**:
  - O **TikTok amplifica o impacto dos atributos criativos**
  - O **LinkedIn apresenta comportamento mais conservador**
  - O **Meta tende a ficar em um nível intermediário**

👉 Portanto, embora existam boas práticas gerais, **a eficácia de cada atributo depende do contexto da plataforma**, sendo fundamental adaptar os criativos para cada ambiente.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# =========================
# 1. PADRÃO VISUAL
# =========================
sns.set_theme(style="whitegrid", context="talk")

FIGSIZE = (10, 5.5)
TITLE_SIZE = 16
LABEL_SIZE = 13
TICK_SIZE = 11
LEGEND_SIZE = 11

def format_ax(ax, title, xlabel, ylabel, legend_title=None, ylim=None):
    ax.set_title(title, fontsize=TITLE_SIZE, pad=14)
    ax.set_xlabel(xlabel, fontsize=LABEL_SIZE)
    ax.set_ylabel(ylabel, fontsize=LABEL_SIZE)
    ax.tick_params(axis="both", labelsize=TICK_SIZE)
    
    if ylim is not None:
        ax.set_ylim(ylim)
    
    if ax.get_legend() is not None:
        ax.legend(title=legend_title, fontsize=LEGEND_SIZE, title_fontsize=LEGEND_SIZE)

def annotate_bars(ax, fmt="{:.1f}", percent=False, y_offset=0.5, min_height=0):
    for p in ax.patches:
        height = p.get_height()
        if height > min_height:
            text = f"{100 * height:.0f}%" if percent else fmt.format(height)
            ax.annotate(
                text,
                (p.get_x() + p.get_width() / 2, height + y_offset),
                ha="center",
                va="bottom",
                fontsize=11
            )

# =========================
# 2. BASES AUXILIARES
# =========================

# Gráfico 1: impacto no klike_score
driver_plot_df = pd.DataFrame({
    "feature": ["Hook", "Hook", "Rosto", "Rosto", "CTA", "CTA"],
    "value": ["Com", "Sem", "Com", "Sem", "Com", "Sem"],
    "metric": [
        df[df["has_hook"] == True]["klike_score"].mean(),
        df[df["has_hook"] == False]["klike_score"].mean(),
        df[df["has_face"] == True]["klike_score"].mean(),
        df[df["has_face"] == False]["klike_score"].mean(),
        df[df["has_cta"] == True]["klike_score"].mean(),
        df[df["has_cta"] == False]["klike_score"].mean(),
    ]
})

# Gráfico 2: probabilidade de alta performance
prob_plot_df = pd.DataFrame({
    "feature": ["Hook", "Hook", "Rosto", "Rosto", "CTA", "CTA"],
    "value": ["Com", "Sem", "Com", "Sem", "Com", "Sem"],
    "metric": [
        df[df["has_hook"] == True]["high_performance"].mean(),
        df[df["has_hook"] == False]["high_performance"].mean(),
        df[df["has_face"] == True]["high_performance"].mean(),
        df[df["has_face"] == False]["high_performance"].mean(),
        df[df["has_cta"] == True]["high_performance"].mean(),
        df[df["has_cta"] == False]["high_performance"].mean(),
    ]
})

# Gráfico 3: top vs bottom performers
top = df[df["klike_score"] >= df["klike_score"].quantile(0.75)]
bottom = df[df["klike_score"] <= df["klike_score"].quantile(0.25)]

comparison_df = pd.DataFrame({
    "grupo": ["Top 25%", "Bottom 25%"],
    "Hook": [top["has_hook"].mean(), bottom["has_hook"].mean()],
    "Rosto": [top["has_face"].mean(), bottom["has_face"].mean()],
    "CTR": [top["ctr"].mean(), bottom["ctr"].mean()],
    "ROAS": [top["roas"].mean(), bottom["roas"].mean()]
})

comparison_long = comparison_df.melt(
    id_vars="grupo",
    var_name="metrica",
    value_name="valor"
)

# =========================
# 3. VISUALIZAÇÃO 1
# Impacto dos atributos no Klike Score
# =========================
plt.figure(figsize=FIGSIZE)
ax = sns.barplot(data=driver_plot_df, x="feature", y="metric", hue="value")
format_ax(
    ax,
    title="Impacto dos atributos criativos no Klike Score",
    xlabel="Atributo",
    ylabel="Klike Score médio",
    legend_title="Presença"
)
annotate_bars(ax, fmt="{:.1f}", y_offset=0.6, min_height=0)
plt.tight_layout()
plt.show()

# =========================
# 4. VISUALIZAÇÃO 2
# Probabilidade de alta performance
# =========================
plt.figure(figsize=FIGSIZE)
ax = sns.barplot(data=prob_plot_df, x="feature", y="metric", hue="value")
format_ax(
    ax,
    title="Probabilidade de alta performance por atributo criativo",
    xlabel="Atributo",
    ylabel="Probabilidade",
    legend_title="Presença",
    ylim=(0, 0.8)
)
annotate_bars(ax, percent=True, y_offset=0.015, min_height=0)
plt.tight_layout()
plt.show()

# =========================
# 5. VISUALIZAÇÃO 3
# Top vs Bottom performers
# =========================
plt.figure(figsize=FIGSIZE)
ax = sns.barplot(data=comparison_long, x="metrica", y="valor", hue="grupo")
format_ax(
    ax,
    title="Top 25% vs Bottom 25%: diferenças entre campanhas",
    xlabel="Métrica",
    ylabel="Valor médio / proporção",
    legend_title="Grupo"
)
annotate_bars(ax, fmt="{:.2f}", y_offset=0.03, min_height=0)
plt.tight_layout()
plt.show()

# =========================
# 6. VISUALIZAÇÃO 4
# Impacto do hook por plataforma
# =========================
plt.figure(figsize=FIGSIZE)
ax = sns.barplot(data=df, x="platform", y="klike_score", hue="has_hook", errorbar=None)
format_ax(
    ax,
    title="Impacto do hook no Klike Score por plataforma",
    xlabel="Plataforma",
    ylabel="Klike Score médio",
    legend_title="Has Hook"
)
annotate_bars(ax, fmt="{:.1f}", y_offset=0.6, min_height=0)
plt.tight_layout()
plt.show()

# =========================
# 7. VISUALIZAÇÃO 5
# Densidade de texto
# =========================
plt.figure(figsize=FIGSIZE)
ax = sns.barplot(data=df, x="text_density", y="klike_score", order=["low", "medium", "high"], errorbar=None)
format_ax(
    ax,
    title="Impacto da densidade de texto no Klike Score",
    xlabel="Densidade de texto",
    ylabel="Klike Score médio"
)
annotate_bars(ax, fmt="{:.1f}", y_offset=0.6, min_height=0)
plt.tight_layout()
plt.show()
```
